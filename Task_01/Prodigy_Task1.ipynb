```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/home/adi-pandey/Downloads/house_pricesP1.csv')
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['square_footage','bedrooms','bathrooms']],df.price)
```

```python
reg.coef_
#m1 m2 m3
```

```python
reg.intercept_
#b
```

```python
reg.predict(pd.DataFrame({
    'square_footage': [3000],
    'bedrooms': [6],
    'bathrooms': [3]}))
```
