# Customer Segmentation using K-Means Clustering
### Internship Task
**Objective:** Group retail customers based on their purchase history using K-Means clustering.

**Dataset:** `Retail_Customer_Segmentation.csv`

## Step 1: Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

plt.style.use('default')
```

## Step 2: Load Dataset

```
df = pd.read_csv("Retail_Customer_Segmentation.csv")
df.head()
```

## Step 3: Basic Information

```
print("Shape:", df.shape)
df.info()
df.describe()
```

## Step 4: Missing Values

```
df.isnull().sum()
```

## Step 5: Exploratory Data Analysis

```
df.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

## Step 6: Correlation Matrix

```
corr=df.drop(columns=["CustomerID"]).corr()

plt.figure(figsize=(7,6))
plt.imshow(corr,cmap='coolwarm')
plt.xticks(range(len(corr.columns)),corr.columns,rotation=90)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()

corr
```

## Step 7: Feature Selection

```
X=df.drop(columns=["CustomerID"])
X.head()
```

## Step 8: Feature Scaling

```
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled[:5]
```

## Step 9: Elbow Method

```
wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,random_state=42,n_init=10)
    model.fit(X_scaled)
    wcss.append(model.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.grid(True)
plt.show()
```

## Step 10: Silhouette Score

```
scores=[]
for k in range(2,11):
    model=KMeans(n_clusters=k,random_state=42,n_init=10)
    labels=model.fit_predict(X_scaled)
    scores.append(silhouette_score(X_scaled,labels))

plt.figure(figsize=(6,4))
plt.plot(range(2,11),scores,marker='o')
plt.xlabel("Clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()

for i,s in enumerate(scores,start=2):
    print(f"{i} clusters -> {s:.3f}")
```

## Step 11: Train Final K-Means Model
Using **4 clusters**.

```
kmeans=KMeans(n_clusters=4,random_state=42,n_init=10)
df["Cluster"]=kmeans.fit_predict(X_scaled)
df.head()
```

## Step 12: Visualize Clusters

```
plt.figure(figsize=(8,6))
plt.scatter(df["AnnualIncome_k$"],df["SpendingScore"],c=df["Cluster"])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()
```

## Step 13: Cluster Analysis

```
cluster_summary=df.groupby("Cluster").mean(numeric_only=True)
cluster_summary
```

## Step 14: Save Output

```
df.to_csv("Customer_Segmentation_Output.csv",index=False)
print("Saved successfully!")
```
